```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#Loading the datasets
df = pd.read_csv("tmdb_5000_movies.csv")
df
```

```python
df.head(3)
```

```python
#Selecting important features
selected_features = ["genres","keywords","original_title","overview","tagline","production_companies"]
```

```python
#finding missing values
df.isnull().sum()
```

```python
#filling null values with empty string of selected_features
for feature in selected_features:
    df[feature] = df[feature].fillna(" ")
    
```

```python
#finding missing values
df.isnull().sum()
```

```python
#Marging all 5 selected features
margerd_features = df["genres"]+" "+df["keywords"]+" "+df["original_title"]+" "+df["overview"]+" "+df["tagline"]+" "+df["production_companies"]
margerd_features
```

```python
#Converting text data to numeric data
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
vectorized_features = vectorizer.fit_transform(margerd_features)
vectorized_features = vectorized_features.toarray()
vectorized_features
```

```python
#Getting the simirality scores wih cosine_simirality
from sklearn.metrics.pairwise import cosine_similarity
simirality = cosine_similarity(vectorized_features)
simirality
```

```python
#getting the all the movies name from the original df datasets as a list
movies_name = list(df["original_title"])
movies_name
```

```python
import difflib

#finding close matching movies name from movies_name list according to input_movie name
matched_movies = difflib.get_close_matches(input_movie,movies_name)
matched_movies
```

```python
closed_matched_movie = matched_movies[0]
closed_matched_movie
```

```python
#Getting the index number of closed_matched_movie
index_of_the_movie = df[df["original_title"]==closed_matched_movie].index[0]
index_of_the_movie
```

```python
#Extracting the similarity score according to index_number
similarity_score = list(enumerate(simirality[index_of_the_movie]))
similarity_score
```

# In each tuple,first element is index and second element is similarity_score

```python
#Sorting the movies based on their similarity score(key=lambda x:x[1]) not on index(key=lambda x:x[0])
sorted_similar_movies = sorted(similarity_score,key=lambda x:x[1],reverse=True)
sorted_similar_movies
```

# Movie Recommadation System

```python
#getting the input Movie Name from the users
input_movie = input("Enter a Movie Name :")

#getting the all the movies name from the original df datasets as a list
movies_name = list(df["original_title"])

#finding close matching movies name from movies_name list according to input_movie name
matched_movies = difflib.get_close_matches(input_movie,movies_name)

closed_matched_movie = matched_movies[0]

#Getting the index number of closed_matched_movie
index_of_the_movie = df[df["original_title"]==closed_matched_movie].index[0]

#Extracting the similarity score according to index_number
similarity_score = list(enumerate(simirality[index_of_the_movie]))

#Sorting the movies based on their similarity score not on index
sorted_similar_movies = sorted(similarity_score,key=lambda x:x[1],reverse=True)

#Printing the similar movies based on index from sorted_similar_movies
print("Movies Suggested for you => \n")

#Hare,we start with i = 1,because we do not include frist movie(input movie)
i = 1
for index,score in sorted_similar_movies:
    movie_names_from_index = df[df.index==index].original_title.values[0]
    if i<11:
        print(i,".",movie_names_from_index)
        i+=1
```
